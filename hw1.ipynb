```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y = np.arange(-0.1,1.1,.01)
x = np.ones(len(y))*.8
plt.plot(x,y,'red')
plt.xlabel('$n$', {'fontsize':'large'}, x=0.99)
plt.ylabel('$w$', {'fontsize':'large'}, y=0.99, rotation='horizontal')
plt.xlim([0,1])
plt.xticks([0,0.8,1], ['0', '$\\frac{1}{D}$', '1'])
plt.yticks([])
plt.savefig('fig1.pdf')
```

```python
x = np.arange(0.0, 1.1, .01)
y1 = 1-x
y2 = 1-1.5*x
plt.plot(x,y1,'red')
plt.plot(x,y2,'r--')
plt.plot(.5,.5, 'bo')
plt.annotate('$E$', xy=(.5,.5), xytext=(.51,.51))
plt.plot(.333,.5,'bo')
plt.annotate('$E\'$', xy=(.333,.5), xytext=(.343, .51))
#plt.annotate('$P_1$', xy=(.333,.5), xytext=(.5,.5), arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))
plt.arrow(.85,.1,-(.5-.333),0, color='b')
plt.annotate('$P_1\\uparrow$', xy=(.7,.15), xytext=(.7,.15))
plt.xlabel('$c_1$',fontsize='large', x=0.99)
plt.ylabel('$c_2$', fontsize='large', y=0.99,rotation='horizontal')
plt.xticks([0,.5,.333,1], ['0','$y_1=\\frac{Y_1}{P_1}$','$y_1\'$', '$y_1+\\frac{y_2}{1+r}$'])
plt.yticks([.5, 1], ['$y_2=\\frac{Y_2}{P_2}$', '$y_1(1+r)+y_2$'])
plt.xlim([0,1.2])
plt.ylim([0,1.2])
plt.savefig('fig2.pdf')
```
